```python
import numpy as np
import pandas as pd

# use pandas to extract rainfall inches as a NumPy array
rainfall = pd.read_csv('Seattle2014.csv')['PRCP'].values
inches = rainfall / 254.0  # 1/10mm -> inches
inches.shape

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()  # set plot styles
plt.hist(inches, 30);

days_no_rain = np.count_nonzero(inches == 0)
days_with_rain = 365 - days_no_rain
days_big_rain = np.count_nonzero(inches > 0.5)
days_small_rain = np.count_nonzero(((inches < 0.2) * (inches > 0)))

print("Number days without rain:      ", days_no_rain)
print("Number days with rain:         ", days_with_rain)
print("Days with more than 0.5 inches:", days_big_rain)
print("Rainy days with < 0.2 inches  :", days_small_rain)

median_rainy = np.median(inches[inches > 0])
summer_days = inches[172:262]
median_summer = np.median(summer_days)
max_summer = np.amax(summer_days)

y = np.append(inches[:172],inches[262:])  
median_no_summer = np.median(y)

print("Median precip on rainy days in 2014 (inches):   ", median_rainy)
print("Median precip on summer days in 2014 (inches):  ", median_summer)
print("Maximum precip on summer days in 2014 (inches): ", max_summer)
print("Median precip on non-summer rainy days (inches):", median_no_summer)

```
