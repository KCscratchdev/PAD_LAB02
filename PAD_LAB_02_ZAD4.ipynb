```python
import numpy as np
import pandas as pd

A = [0,3,2,5]
B = [0,3,1,4]

sum = np.add(A, B)
sub = np.subtract(A, B)
aScaled = np.multiply(A, 4)
dotProd = np.dot(A, B)
bSize = np.size(B)

print("Sum = ", sum)
print("Subtract = ", sub)
print("A multiplied by 4 = ", aScaled)
print("Dot product = ", dotProd)
print("B size = ", bSize)



```
