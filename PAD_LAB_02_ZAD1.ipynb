```python
import pandas as pd
import numpy as np

data = pd.read_csv('president_heights.csv')
heights = np.array(data['height(cm)'])

avg = np.mean(heights)
stdeviation = np.std(heights)
min = np.amin(heights)
max = np.amax(heights)

print("Mean height:       ", avg)
print("Standard deviation:", stdeviation)
print("Minimum height:    ", min)
print("Maximum height:    ", max)

twoFive = np.percentile(heights, 25)
sFive = np.percentile(heights, 75)
med = np.median(heights)
print("25th percentile:   ", twoFive)
print("Median:            ", sFive)
print("75th percentile:   ", med)
```
